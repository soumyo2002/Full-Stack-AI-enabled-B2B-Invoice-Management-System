### Q1)Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.


```python
# Visualizing Outliers

import matplotlib.pyplot as plt

# Specify the column containing the 'amount_in_usd' data
amount_column = df['amount_in_usd']

# Create a boxplot to visualize the distribution of the 'amount_in_usd' data
plt.boxplot(amount_column)

# Add labels and title to the plot
plt.xlabel('amount_in_usd')
plt.ylabel('Value')
plt.title('Boxplot of amount_in_usd')

# Display the plot
plt.show()
```

#### Outliers are clearly visible in the graph

```python
# Function to identify outliers using Standard deviation
def outliers(data):
    num = 3
    upper_lim = data.mean () + data.std () * num  
    lower_lim = data.mean () - data.std () * num  
    print("Upper Limit : ", upper_lim, "\nLower Limit : ",lower_lim )

    data = data[(data < upper_lim) & (data > lower_lim)]
    print(data)
```

```python
# displaying upper and lower limit for outliers
outliers(df['amount_in_usd'])
```

```python
# As the data seems to have good amount of outliers its best to replace it with median as mean 
# can get affected by high amount of outliers
upper_lim = 80202.65483141167
lower_lim =-75069.5518797524
outliers = df[(df['amount_in_usd'] > upper_lim) | (df['amount_in_usd'] < lower_lim)].index
median_value = df['amount_in_usd'].median()

# Replace outliers with the median value
df.loc[outliers, 'amount_in_usd'] = median_value
```

```python
# Specify the column containing the 'amount_in_usd' data
amount_column = df['amount_in_usd']

# Create a boxplot to visualize the distribution of the 'amount_in_usd' data
plt.boxplot(amount_column)

# Add labels and title to the plot
plt.xlabel('amount_in_usd')
plt.ylabel('Value')
plt.title('Boxplot of amount_in_usd')

# Display the plot
plt.show()
```

#### Reason for not replacing outliers with:
#### 1) Mean:
#### The data is outlierish in nature and replacing it with mean will introduce a bias as mean is sensitive to outliers and it will shift the overall values towards outliers

#### 2)Mode:
#### Replacing outliers with mode can introduce a bias as it will artificially inflate the frequency of mode further

#### Also it must be noted that outliers can cause data leakage indirectly as we are replacing it with median of entire dataset which also contains data of test dataset but still I am doing it before train test split as outliers for whole data maybe not be an outlier for train or test data,thus it's better to remove it now otherwise later it can cause bias in predictions

```python
# For each customer, this function computes the lagged sales amounts, 
# and then computes the discrepancies between these lagged sales amounts. 
# These variations can offer information about how each customer's sales have changed over time.


def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt
```

```python
# with the following function we can select highly correlated features,it will remove the first 
# feature that is highly correlated with any other other feature 
def correlation(dataset, threshold): 
    col_corr = set() # Set of all the names of correlated columns 
    corr_matrix = dataset.corr() 
    for i in range(len(corr_matrix.columns)):
        for j in range(i): 
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value 
                colname = corr_matrix.columns[i] # getting the name of column 
                col_corr.add(colname)
    return col_corr 
```

#### Reason for this step is that two features highly correlated don't add anything new to the Machine learning model,thus it's better to remove them as too many features can hamper prediction accuracy
